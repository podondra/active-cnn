# True positive estimation

Because there are no true labels in LAMOST data, the accuracy need to be estimated. Here I use *confidence interval* to estimate the uncertainty in the estimate. I provide the 95% lower bound.

```python
import numpy as np
import pandas as pd
from active_cnn.plot import preview_lamost_spectrum
from active_cnn.performance import lower_confidence_limit
```

```python
# seed to get reproducible result: random.org
np.random.seed(50)
```

```python
df = pd.read_csv('data/run-9-11-2018/candidates-double-peak.csv')
shuffled_filenames = (f for f in df.sample(frac=1).filename.values)
sample = []
```

```python
preview_lamost_spectrum(next(shuffled_filenames), 2)
```

```python
# add one to the sample list if classification is correct
sample.append(1)
```

```python
# else add zero
sample.append(0)
```

```python
lower_confidence_limit(sample)
```

```python
# show what decisions I did
print(sample)
```
